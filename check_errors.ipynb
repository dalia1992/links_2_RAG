```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
report = pd.read_csv('reporte_procesamiento.csv')
```

```python
print(len(report))
display(report.groupby('Estado').count())

display(report.groupby('Estado').count()/len(report)*100)

```

```python
success = report[report['Estado']=='EXITO']
print(f"Tiempo promedio de éxito: {round(success['Tiempo_Segundos'].mean(),2)}")
print(f"Tiempo máximo de éxito: {round(success['Tiempo_Segundos'].max(),2)}")
print(f"Tiempo mínimo de éxito: {round(success['Tiempo_Segundos'].min(),2)}")

sns.displot(success, x='Tiempo_Segundos')
plt.title('Distribución de tiempo por casos exitosos')

```

```python
print(f"Access denied: {sum(['access denied' in x for x in report['Detalle_Error'].dropna()])}")

print(f"Page not found: {sum(['not found' in x for x in report['Detalle_Error'].dropna()])}")

print(f"Captcha required: {sum(['captcha' in x for x in report['Detalle_Error'].dropna()])}")

print(f"No se hizo retrieval de contenido suficiente: {sum(['Not enough content retrieved' in x for x in report['Detalle_Error'].dropna()])}")

```
